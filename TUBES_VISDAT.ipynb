```python
import pandas as pd

df = pd.read_csv('/content/level3_road_cong.csv')
level_1 = pd.read_excel('/content/Level_1.xlsx')
```

```python
import pandas as pd
import plotly.graph_objects as go

years = [str(y) for y in range(2025, 2051)]

congestion_data = df[df['co-benefit_type'] == 'congestion'][years].astype(float).sum()
road_safety_data = df[df['co-benefit_type'] == 'road_safety'][years].astype(float).sum()

congestion_cumulative = congestion_data.cumsum()
road_safety_cumulative = road_safety_data.cumsum()

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=years,
    y=congestion_cumulative.values,
    mode='lines+markers',
    name='Congestion Benefits',
    line=dict(color='#2E86AB', width=3),
    marker=dict(size=6)
))

fig.add_trace(go.Scatter(
    x=years,
    y=road_safety_cumulative.values,
    mode='lines+markers',
    name='Road Safety Benefits',
    line=dict(color='#E94F37', width=3),
    marker=dict(size=6)
))

fig.update_layout(
    title='<b>Cumulative Economic Benefits 2025-2050</b><br><sup>Long-term Value of Climate Policy Investment</sup>',
    xaxis_title='Year',
    yaxis_title='Cumulative Economic Value (£M)',
    template='plotly_white',
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=-0.3,
        xanchor="center",
        x=0.5
    ),
    margin=dict(b=100),
    hovermode='x unified',
    height=500
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px
import statsmodels

congestion = df[df['co-benefit_type'] == 'congestion'].groupby('small_area')['sum'].sum().reset_index()
congestion.columns = ['small_area', 'congestion_benefit']

road_safety = df[df['co-benefit_type'] == 'road_safety'].groupby('small_area')['sum'].sum().reset_index()
road_safety.columns = ['small_area', 'road_safety_benefit']

merged = pd.merge(congestion, road_safety, on='small_area')
sample = merged.sample(n=min(5000, len(merged)), random_state=42)

fig = px.scatter(
    sample,
    x='congestion_benefit',
    y='road_safety_benefit',
    opacity=0.5,
    color_discrete_sequence=['#3498DB'],
    trendline='ols'
)

fig.update_layout(
    title='<b>Congestion vs Road Safety Benefits</b><br><sup>Synergy Between Efficiency & Safety (Sample of 5,000 Areas)</sup>',
    xaxis_title='Congestion Reduction Benefits (£)',
    yaxis_title='Road Safety Benefits (£)',
    template='plotly_white',
    height=500
)

fig.update_traces(marker=dict(size=6))

fig.show()
```

```python
import geopandas as gpd
import plotly.express as px
import json

gdf = gpd.read_file("/content/small_areas_british_grid.shp")
gdf['geometry'] = gdf['geometry'].simplify(tolerance=0.01, preserve_topology=True)
if gdf.crs != "EPSG:4326": gdf = gdf.to_crs("EPSG:4326")

choropleth_df = level_1[['small_area', 'road_safety', 'congestion']].copy()
choropleth_df['total_benefit'] = choropleth_df['road_safety'] + choropleth_df['congestion']
gdf_merged = gdf.merge(choropleth_df, on="small_area", how="left")
gdf_merged['total_benefit'] = gdf_merged['total_benefit'].fillna(0)

fig_map = px.choropleth_mapbox(
    gdf_merged,
    geojson=json.loads(gdf_merged.to_json()),
    locations='small_area',
    color='total_benefit',
    featureidkey="properties.small_area",
    mapbox_style="carto-positron",
    zoom=4.5,
    center={"lat": 54.5, "lon": -2.5},
    color_continuous_scale="YlGnBu",
    range_color=[gdf_merged['total_benefit'].min(), gdf_merged['total_benefit'].max()]
)

fig_map.update_layout(
    margin={"r":0,"t":40,"l":0,"b":0},
    height=600,
    title='<b>Where Climate Action Saves Time and Lives</b> Total Co-Benefit Breakdown',
)

fig_map.show()
```

```python
import pandas as pd
import plotly.express as px

df['sum'] = pd.to_numeric(df['sum'], errors='coerce')

pathways = df.groupby(['damage_pathway', 'damage_type'])['sum'].sum().abs().reset_index()
pathways.columns = ['Damage Pathway', 'Type', 'Total Value']

pathway_labels = {
    'time_saved': 'Time Saved',
    'reduced_mortality': 'Reduced Mortality',
    'society': 'Society'
}
pathways['Pathway Label'] = pathways['Damage Pathway'].map(pathway_labels)

fig = px.bar_polar(
    pathways,
    r='Total Value',
    theta='Pathway Label',
    color='Type',
    color_discrete_map={'health': '#E94F37', 'non-health': '#2E86AB'},
    template='plotly_white'
)

fig.update_layout(
    title='<b>Damage Pathways Breakdown</b><br><sup>Health vs Non-Health Economic Impact by Category</sup>',
    height=500,
    polar=dict(radialaxis=dict(showticklabels=True, tickangle=45)),
    legend=dict(orientation='h', y=-0.15)
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

mortality = df[df['damage_pathway'] == 'reduced_mortality'].copy()
years = [str(y) for y in range(2025, 2051)]

for col in years:
    mortality[col] = pd.to_numeric(mortality[col], errors='coerce')

yearly_mortality = mortality[years].sum()
VPF = 2.1
lives_saved = yearly_mortality / VPF

cumulative_lives = lives_saved.cumsum()
total_lives = lives_saved.sum()

fig = go.Figure()

colors = ['#27AE60' if v > 0 else '#E74C3C' for v in lives_saved.values]

fig.add_trace(go.Bar(
    x=years,
    y=lives_saved.values,
    marker_color=colors,
    name='Annual Lives',
    text=[f'{v:.1f}' if abs(v) > 1 else '' for v in lives_saved.values],
    textposition='outside'
))

fig.add_trace(go.Scatter(
    x=years,
    y=cumulative_lives.values,
    mode='lines+markers',
    name='Cumulative Lives',
    line=dict(color='#8E44AD', width=3),
    yaxis='y2'
))

fig.update_layout(
    title=f'<b>🧑‍🤝‍🧑 Statistical Lives Saved via Reduced Mortality</b><br><sup>UK DfT VPF Standard (£2.1M/life) | Total: {total_lives:.0f} lives | Green=Saved, Red=At Risk</sup>',
    xaxis_title='Year',
    yaxis_title='Annual Lives Saved',
    yaxis2=dict(title='Cumulative Lives', overlaying='y', side='right', showgrid=False),
    template='plotly_white',
    height=500,
    legend=dict(orientation='h', y=-0.15),
    bargap=0.3
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

lives_df = df[df['damage_pathway'] == 'reduced_mortality'].copy()
years = [str(y) for y in range(2025, 2051)]
lives_saved = []
for year in years:
    lives_df[year] = pd.to_numeric(lives_df[year], errors='coerce')
    positive_sum = lives_df[lives_df[year] > 0][year].sum()
    lives_saved.append(positive_sum)

fig = go.Figure()

max_val = max(lives_saved)
normalized = [int((v / max_val) * 10) for v in lives_saved]

key_years = ['2025', '2030', '2035', '2040', '2045', '2050']
key_indices = [years.index(y) for y in key_years]
key_values = [lives_saved[i] for i in key_indices]
key_norm = [normalized[i] for i in key_indices]

for i, (year, val, norm) in enumerate(zip(key_years, key_values, key_norm)):
    people = '👤' * max(1, norm)
    fig.add_trace(go.Bar(
        y=[year],
        x=[val],
        orientation='h',
        marker_color=['#2ecc71', '#27ae60', '#1abc9c', '#16a085', '#3498db', '#2980b9'][i],
        text=f'{people} {val:.1f}',
        textposition='inside',
        textfont=dict(size=12),
        showlegend=False
    ))

fig.update_layout(
    title=f'🏥 <b>Lives Saved Pictogram</b><br><sub>Reduced Mortality Benefits by Year</sub>',
    xaxis_title='Lives Saved (Units)',
    yaxis_title='Year',
    plot_bgcolor='white',
    height=400,
    width=450,
    bargap=0.3
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px
import numpy as np

df['region'] = df['small_area'].str[:1]
region_names = {'E': 'England', 'W': 'Wales', 'S': 'Scotland', 'N': 'N. Ireland'}
df['region_name'] = df['region'].map(region_names).fillna('Other')

df['sum'] = pd.to_numeric(df['sum'], errors='coerce')
pivot_df = df.groupby(['region_name', 'co-benefit_type', 'damage_pathway'])['sum'].sum().reset_index()

pivot_df['category'] = pivot_df['co-benefit_type'] + '\n' + pivot_df['damage_pathway']

heatmap_data = pivot_df.pivot_table(
    index='region_name',
    columns='category',
    values='sum',
    aggfunc='sum'
).fillna(0)

fig = px.imshow(
    heatmap_data.values,
    x=heatmap_data.columns.tolist(),
    y=heatmap_data.index.tolist(),
    color_continuous_scale='RdYlGn',
    aspect='auto',
    text_auto='.1f'
)

fig.update_layout(
    title=f'🗺️ <b>Regional Heatmap</b><br><sub>Co-Benefits by Region & Pathway</sub>',
    xaxis_title='Benefit Category',
    yaxis_title='Region',
    height=400,
    width=450,
    coloraxis_colorbar=dict(title='Sum')
)

fig.update_xaxes(tickangle=45, tickfont=dict(size=9))

fig.show()
```
