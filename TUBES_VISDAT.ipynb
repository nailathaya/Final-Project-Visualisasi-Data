```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load shapefile
gdf = gpd.read_file(f"/content/drive/MyDrive/Data Visualisation Competition 2025/shapefile/small_areas_british_grid.shp")
```

# Choropleth Map

```python
# Prepare dataset
choropleth_df = level_1[['small_area', 'road_safety', 'congestion']].copy()
choropleth_df['total_benefit'] = choropleth_df['road_safety'] + choropleth_df['congestion']

# Join with shapefile
gdf_merged = gdf.merge(choropleth_df, on="small_area", how="left")

# Plot
fig, ax = plt.subplots(figsize=(12, 14))
gdf_merged.plot(column='total_benefit',
                cmap='viridis',
                legend=True,
                linewidth=0,
                ax=ax)

ax.set_title("Where Climate Action Saves Time and Lives\nTotal Co-Benefit: Road Safety + Congestion",
             fontsize=16)
ax.axis('off')
plt.show()
```

# Scatter Plot

```python
# Merge lookup data
scatter_df = level_1[['small_area', 'road_safety', 'congestion']].merge(
    lookups[['small_area', 'population', 'local_authority']],
    on='small_area',
    how='left'
)

# Remove missing populations
scatter_df = scatter_df.dropna(subset=['population'])
```

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(
    data=scatter_df,
    x="congestion",
    y="road_safety",
    size="population",
    hue="local_authority",
    alpha=0.6,
    legend=False
)

plt.title("Safer Roads Come with Smoother Traffic", fontsize=16)
plt.xlabel("Congestion Benefit (million GBP)")
plt.ylabel("Road Safety Benefit (million GBP)")
plt.grid(True, alpha=0.3)
plt.show()
```

# Line Chart

```python
# Filter per co-benefit
cong_df = level_3[
    (level_3['co-benefit_type'] == 'congestion') &
    (level_3['damage_pathway'] == 'time_saved')
]

rs_df = level_3[
    (level_3['co-benefit_type'] == 'road_safety') &
    (level_3['damage_pathway'].isin(['reduced_mortality', 'society']))
]
```

```python
# Extract year columns
year_cols = [y for y in range(2025, 2051)]

# Sum congestion time_saved by year
cong_yearly = cong_df[year_cols].sum()

# Sum road safety (two pathways) by year
rs_yearly = rs_df[year_cols].sum()
```

```python
plt.figure(figsize=(14, 7))

plt.plot(cong_yearly.index, cong_yearly.values,
         label='Congestion Reduction (time_saved)', linewidth=2.5)

plt.plot(rs_yearly.index, rs_yearly.values,
         label='Road Safety (reduced_mortality + society)', linewidth=2.5)

plt.title("The Growing Benefits of Climate Action (2025â€“2050)", fontsize=18)
plt.xlabel("Year")
plt.ylabel("Economic Benefit (million GBP, NPV 2025)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
